# **Importing Library**

```python
import pandas as pd
```

# **Importing the Dataset**

```python
car=pd.read_csv("CarsData.csv")
```

Checking the Shape of the Dataset

```python
car.shape
```

# **Cleaning the data**

Checking for Null values

```python
car.isnull().sum()
```

Replacing the null values with the mean of specific column

```python
car['Cylinders'].fillna(car['Cylinders'].mean(),inplace=True)
```

Counting the occurences of each make

```python
car['Make'].value_counts()
```

Filtering based on certain value

```python
car[car['Origin'].isin(['Asia','Europe'])]
```

Removing certain values based on a condition

```python
car[~(car['Weight']>4000)]
```

Increasing the values in a column by 5

```python
car['MPG_City']=car['MPG_City'].apply(lambda x:x+5)
```
